```python
import requests
from bs4 import BeautifulSoup

#Create a variable to save the URL
url ='https://en.wikipedia.org/wiki/Comma-separated_values'

# Specify the URL and request with get()
r = requests.get(url)

# Determine a connection
if r.status_code == 200:
    html_doc = r.text
    
print(r)
```

```python
# Set BeautifulSoup object
soup = BeautifulSoup(html_doc)
print(soup.prettify())
```

```python
# Find the html element and then extract it
tables = soup.find('table',{'class':'wikitable'})
tables
```

```python
# Extract all elements with <td> which is the car data
cars = tables.findAll('td')

cars
```
